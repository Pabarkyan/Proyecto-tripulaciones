```python
for i in range(9, 26):
    columna = f'Unnamed: {i}'
    data.drop(columns=[columna], inplace=True, errors='ignore')

data.head()
```

```python
# Parece que quedan algunos null que deben ser que la tarifa no tiene informacion de alguna columna
data = data[data['compania'].isna() != True]
data.info()
```

```python
data[data.isnull().any(axis=1)]
```

```python
data = data.dropna()
```

```python
# Metemos dentro de los intervalos los valores individuales de potencia contratada
print(data['potencia_contratada'].dtypes)

data['potencia_contratada'].unique() # valores a usar: 'Entre 10 y 15', '>15', '<=15', '<=10'
```

```python
def intervalos(valor):
    try:
        valor = float(valor)
        if valor <= 10:
            return '≤10'
        elif (valor > 10) and (valor < 15):
            return 'Entre 10 y 15'
        elif valor == 15:
            return '≤15'
        else:
            return '>15'

    except:
        return valor

data['potencia_contratada'] = data['potencia_contratada'].apply(lambda x: intervalos(x))
```
